```python
# r= (no of cases on next day- no of cases on current day)/(no of cases on current day) )

r = 0.0
start_date = df2.index[df2['Date']=='2020-03-04'][0]
end_date= df2.index[df2['Date']=='2020-03-21'][0]
for i in range(start_date, end_date):
    current_day = df2.loc[i]['Total_cases']
    next_day= df2.loc[i+1]['Total_cases']
    daily_rate=(next_day-current_day)/(current_day)
r = daily_rate
```
